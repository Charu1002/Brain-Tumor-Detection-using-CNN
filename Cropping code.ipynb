```python
import cv2
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
image = cv2.imread(r'./data/brain_tumor_dataset/no/10 no.jpg')
```

```python
plt.imshow(image)
```

```python
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
lower = np.array([30,30,30])
higher = np.array([250,250,250])
```

```python
mask = cv2.inRange(img,lower,higher)
mask.shape
```

```python
plt.imshow(mask,'gray')
```

```python
cnts,_ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```

```python
cntimage = cv2.drawContours(img,cnts,-1,255,3)
```

```python
plt.imshow(cntimage)
```

```python
c = max(cnts, key=cv2.contourArea)
```

```python
x,y,w,h = cv2.boundingRect(c)
```

```python
cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),1)
```

```python
plt.imshow(img)
```

```python
cropped_image = image[y:y+h,x:x+w]
```

```python
plt.imshow(cropped_image)
```

```python
image_width, image_height = 240,240
im = cv2.resize(cropped_image, dsize=(image_width, image_height), interpolation=cv2.INTER_CUBIC)
im = im / 255
```

```python
plt.imshow(im)
```
